```python
# Name: Jelle Stoffels
# Section: Programmeren IK
# Date: 6 October 2014    
# hanoi.py

# This program enables the user to play a game of Hanoi.

import pygame, sys, time
from pygame.locals import *

#
# Defining variables and parameters
#

# Initialise game
pygame.init()
screen = pygame.display.set_mode((900, 600))
pygame.display.set_caption('Towers of Hanoi')

# Some bureaucracy
stone_selected = False
animating = False
invalid = False
stones = []
count = 0
scene = 1
number_of_stones = 3
left = 1
invalid = False

# Fill background
background = pygame.Surface(screen.get_size())
background = background.convert()

# colors
brown = (150,140,120)
brown2 = (50, 40, 20)
grey1 = (240,240,255)
grey2 = (200,200,220)
grey3 = (160,160,180)
grey4 = (120,120,140)
grey5 = (80,80,100)
grey6 = (60,60,80)
grey7 = (40,40,60)
grey8 = (20,20,40)
grey9 = (10,10,30)
white = (240,240,245)
red = (255,100,100)
red2 = (220, 65, 65)

#
# Functions: drawing screens
#

# Draws the first screen: the start menu.
def draw_screen1():
    
    self.g.draw_menu(self)
    
# Draws the second screen: the accual game.
def draw_screen2():

	global back_button

	background.fill((150, 120, 120))
	screen.fill((50,50,50))

	# Shapes
	pygame.draw.line(screen, brown, (120, 500), (780, 500), 101) # Big block
	
	tower_1 = pygame.draw.line(screen, brown, (250, 450), (250, 250), 16)	# tower 1, 2 and 3
	tower_2 = pygame.draw.line(screen, brown, (450, 450), (450, 250), 16)
	tower_3 = pygame.draw.line(screen, brown, (650, 450), (650, 250), 16)
	
	pygame.draw.circle(screen, brown, (251,250), 8, 0) # circles on towers
	pygame.draw.circle(screen, brown, (451,250), 8, 0)
	pygame.draw.circle(screen, brown, (651,250), 8, 0)

	stone4 = pygame.draw.rect(screen, grey5, stones[0]) # stones
	if number_of_stones > 1:
		stone3 = pygame.draw.rect(screen, grey4, stones[1])
	if number_of_stones > 2:
		stone2 = pygame.draw.rect(screen, grey3, stones[2])
	if number_of_stones > 3:
		stone1 = pygame.draw.rect(screen, grey2, stones[3])
	if number_of_stones > 4:
		stone0 = pygame.draw.rect(screen, grey1, stones[4])

	back_button = pygame.draw.rect(screen, red, (20,20,120,40))

	# Text
	fontObj = pygame.font.Font('freesansbold.ttf', 21)
	textSurfaceObj = fontObj.render('Go back', True, white)
	textRectObj = textSurfaceObj.get_rect()
	textRectObj.center = (80, 40)
	screen.blit(textSurfaceObj, textRectObj)

	fontObj = pygame.font.Font('freesansbold.ttf', 32)
	textSurfaceObj = fontObj.render('Towers of Hanoi', True, brown2)
	textRectObj = textSurfaceObj.get_rect()
	textRectObj.center = (450, 500)
	screen.blit(textSurfaceObj, textRectObj)

	fontObj = pygame.font.Font('freesansbold.ttf', 21)
	textSurfaceObj = fontObj.render('Turns: ' + str(count), True, white)
	textRectObj = textSurfaceObj.get_rect()
	textRectObj.center = (800, 50)
	screen.blit(textSurfaceObj, textRectObj)	

	if invalid == True:
		fontObj = pygame.font.Font('freesansbold.ttf', 21)
		textSurfaceObj = fontObj.render("Invalid move. Check the instructions"\
		" if you don't know the rules." , True, white)
		textRectObj = textSurfaceObj.get_rect()
		textRectObj.center = (450, 200)
		screen.blit(textSurfaceObj, textRectObj)

# Draws the third screen: the instructions.
def draw_screen3():

	global back_button_2

	background.fill((150, 120, 120))
	screen.fill((50,50,50))

	# Shapes
	back_button_2 = pygame.draw.rect(screen, red, (20,20,120,40))

	# Text
	fontObj = pygame.font.Font('freesansbold.ttf', 64)
	textSurfaceObj = fontObj.render('Instructions', True, white)
	textRectObj = textSurfaceObj.get_rect()
	textRectObj.center = (450, 100)
	screen.blit(textSurfaceObj, textRectObj)

	fontObj = pygame.font.Font('freesansbold.ttf', 21)
	textSurfaceObj = fontObj.render('Go back', True, white)
	textRectObj = textSurfaceObj.get_rect()
	textRectObj.center = (80, 40)
	screen.blit(textSurfaceObj, textRectObj)

	fontObj = pygame.font.Font('freesansbold.ttf', 21)
	textSurfaceObj = fontObj.render('The goal of this game is to stack all'\
	' of the stones onto the second or third tower.', True, white)
	textRectObj = textSurfaceObj.get_rect()
	textRectObj.center = (450, 280)
	screen.blit(textSurfaceObj, textRectObj)

	fontObj = pygame.font.Font('freesansbold.ttf', 21)
	textSurfaceObj = fontObj.render('Drag and drop the stones,'\
	' smaller stones may only be placed onto bigger stones.', True, white)
	textRectObj = textSurfaceObj.get_rect()
	textRectObj.center = (450, 300)
	screen.blit(textSurfaceObj, textRectObj)

	fontObj = pygame.font.Font('freesansbold.ttf', 42)
	textSurfaceObj = fontObj.render('Good luck!', True, white)
	textRectObj = textSurfaceObj.get_rect()
	textRectObj.center = (450, 380)
	screen.blit(textSurfaceObj, textRectObj)

#
# Functions: game processing
#

# Makes a list of the 3 towers and stacks 
# a certain number of stone onto the first tower,
# depending on the user's choice.
def starting_state(number_of_stones):
	global starting_stones
	global towers
	global tower1
	global tower2
	global tower3

	starting_stones = []
	for n in range(number_of_stones):
		starting_stones.append(n)

	towers = [starting_stones,[],[]]

	tower1 = towers[0]
	tower2 = towers[1]
	tower3 = towers[2]
	
	if number_of_stones > 0:
		stones.append(Rect(160, 430, 180, 20))
	if number_of_stones > 1:
		stones.append(Rect(170, 410, 160, 20))
	if number_of_stones > 2:
		stones.append(Rect(180, 390, 140, 20))
	if number_of_stones > 3:
		stones.append(Rect(190, 370, 120, 20))
	if number_of_stones > 4:
		stones.append(Rect(200, 350, 100, 20))

	stones.reverse()

# Identifies wich button the user clicked on in the start menu.
def identify_button1(x,y):

	if play_button.collidepoint(event.pos):
		button = 'play'
	
	elif stone_button_down.collidepoint(event.pos):
		button = 'down'

	elif stone_button_up.collidepoint(event.pos):
		button = 'up'

	elif instructions_button.collidepoint(event.pos):
		button = 'instructions'

	elif quit_button.collidepoint(event.pos):
		button = 'quit'

	else:
		button = 'none'

	return button

# Identifies the source tower, depending on where the user clicked.
def identify_source_tower():
	if x in range(154, 335) and y in range(235, 450):		
		source_tower = tower1

	elif x in range(340, 550) and y in range(235, 450):
		source_tower = tower2

	elif x in range(560, 750) and y in range(235, 450):
		source_tower = tower3

	elif x in range(0, 900) and y in range(0, 600):
		source_tower = []
	return source_tower

# Identifies the destination tower, depending on where the user declicked.
def identify_dest_tower():
	if x in range(154, 335) and y in range(100, 450):		
		dest_tower = tower1

	elif x in range(340, 550) and y in range(100, 450):
		dest_tower = tower2

	elif x in range(560, 750) and y in range(100, 450):
		dest_tower = tower3

	elif x in range(0, 900) and y in range(0, 600):
		dest_tower = source_tower

	return dest_tower

# Subtracts the smallest stone from the source tower and returns it.
def subtract_stone(source_tower):
	global player_stone	
	player_stone = source_tower.pop(0)

# Checks if the move is valid.
def is_valid_turn(player_stone, dest_tower, source_tower):
	if dest_tower == []:
		return True
	elif player_stone < dest_tower[0]:
		return True	
	elif player_stone > dest_tower[0]:
		return False

# Initiates the add_stone or invalid_move function, depending on the validity of the move.
def move_stone():
	if is_valid_turn(player_stone,	dest_tower,	source_tower) == True:
		add_stone(dest_tower, player_stone)
	if is_valid_turn(player_stone,	dest_tower,	source_tower) == False:
		invalid_move(player_stone, source_tower)

# Returns the stone to the source tower when a move is not valid.
def invalid_move(player_stone, source_tower):
	global invalid
	source_tower.append(player_stone)
	source_tower.sort()
	drop_stone(source_tower)
	invalid = True

# Adds the stone to the destination tower when a move is valid.
def add_stone(dest_tower, player_stone):		
	global invalid
	dest_tower.append(player_stone)
	dest_tower.sort()
	drop_stone(dest_tower)
	invalid = False

# Animates the stone dropping onto a tower when the user releases it.
def drop_stone(dest_tower):

	global animating
	global animating_stone
	global animation_destination_y
	
	heigth = 450 - (len(dest_tower)*20)

	if dest_tower == tower1:
		stones[player_stone].center = (250, heigth-100)
		animating = True
		animating_stone = player_stone
		animation_destination_y = heigth	

	if dest_tower == tower2:
		stones[player_stone].center = (450, heigth-100)
		animating = True
		animating_stone = player_stone
		animation_destination_y = heigth

	if dest_tower == tower3:
		stones[player_stone].center = (650, heigth-100)
		animating = True
		animating_stone = player_stone
		animation_destination_y = heigth

# Counts the turn, if the turn is valid and accualy moves a stone.
def count_turn(count):
	if source_tower != dest_tower:
		if invalid == False:
			count += 1
	return count

# Displays a 'You win!' message on the second screen with the number of turns it took.
def game_won(count):

	fontObj = pygame.font.Font('freesansbold.ttf', 64)
	textSurfaceObj = fontObj.render('You win!', True, red)
	textRectObj = textSurfaceObj.get_rect()
	textRectObj.center = (450, 100)
	screen.blit(textSurfaceObj, textRectObj)

	if count == 1:

		fontObj = pygame.font.Font('freesansbold.ttf', 24)
		textSurfaceObj = fontObj.render('You only needed ' + str(count) + ' turn.', True, white)
		textRectObj = textSurfaceObj.get_rect()
		textRectObj.center = (450, 150)
		screen.blit(textSurfaceObj, textRectObj)
	
	elif count < 35:

		fontObj = pygame.font.Font('freesansbold.ttf', 24)
		textSurfaceObj = fontObj.render('You only needed ' + str(count) + ' turns.', True, white)
		textRectObj = textSurfaceObj.get_rect()
		textRectObj.center = (450, 150)
		screen.blit(textSurfaceObj, textRectObj)	

	elif count > 34:

		fontObj = pygame.font.Font('freesansbold.ttf', 24)
		textSurfaceObj = fontObj.render('But it took you ' + str(count) + ' turns.', True, white)
		textRectObj = textSurfaceObj.get_rect()
		textRectObj.center = (450, 150)
		screen.blit(textSurfaceObj, textRectObj)

	pygame.draw.rect(screen, (50,50,50) , (700, 0, 200, 200))

#
# Main game loop
#

while True:

	(x,y) = pygame.mouse.get_pos()	
	
	# Animations:
	if animating:
		if stones[animating_stone].y != animation_destination_y:
			stones[animating_stone].y += 10
		else:
			animating = False
			animating_stone = False
			animation_destination_y = False

	# Drawing screens:
	if scene == 1:
		draw_screen1()

	elif scene == 2:
		draw_screen2()

		if len(tower2) == number_of_stones:
			game_won(count)
		if len(tower3) == number_of_stones:
			game_won(count)

	elif scene == 3:	
		draw_screen3()

	# Event loop for each scene:
	for event in pygame.event.get():

		if event.type == QUIT:
			pygame.quit()
			sys.exit()
		
		elif scene == 1:

			if event.type == pygame.MOUSEBUTTONDOWN and event.button == left:
				button = identify_button1(x,y)
				if button == 'play': 
					starting_state(number_of_stones)
					scene = 2
				elif button == 'down':
					if number_of_stones > 1:
						number_of_stones = number_of_stones - 1
				elif button == 'up':
					if number_of_stones < 5:
						number_of_stones = number_of_stones + 1
				elif button == 'instructions':
					scene = 3
				elif button == 'quit':
					pygame.quit()
					sys.exit()					

		elif scene == 2:
			
			if event.type == pygame.MOUSEBUTTONDOWN and event.button == left:
				if back_button.collidepoint(event.pos):
					count = 0
					scene = 1

				if x in range(154, 750) and y in range(235, 450): 
					source_tower = identify_source_tower()
					
					if source_tower != []:
						stone_selected = True
						subtract_stone(source_tower)
							
			elif event.type == pygame.MOUSEBUTTONUP and event.button == left:
				try:
					player_stone
				except NameError:
					pass
				else:					
					dest_tower = identify_dest_tower()
					move_stone()
					count = count_turn(count)
					del player_stone

					stone_selected = False

			# Sticks the stone to the mouse when dragged.
			elif event.type == MOUSEMOTION and stone_selected:
				stones[player_stone].center = (x,y)

		elif scene == 3:

			if event.type == pygame.MOUSEBUTTONDOWN and event.button == left:
				if back_button_2.collidepoint(event.pos):
					scene = 1
		
	pygame.display.update()
	pygame.time.wait(10)
```
